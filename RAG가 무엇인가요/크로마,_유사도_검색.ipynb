```python
from google.colab import files
uploaded = files.upload()
```

```python
from langchain.document_loaders import PyPDFLoader

# PDF 가져오기
loaders = [PyPDFLoader(filename) for filename in uploaded.keys()]
docs = []
for loader in loaders:
    docs.extend(loader.load())
```

```python
# Splitter를 이용한 청킹
from langchain.text_splitter import RecursiveCharacterTextSplitter
from openai import OpenAI
client = OpenAI(
  api_key="sk-"
)

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1500,
    chunk_overlap = 150
)

#splitter를 이용한 문서 청킹
splits = text_splitter.split_documents(docs)
print(splits)
```

```python
from langchain.vectorstores import Chroma
from langchain.embeddings.openai import OpenAIEmbeddings
#임베딩 모델 생성
embedding = OpenAIEmbeddings(model="text-embedding-ada-002", openai_api_key="sk-")
```

```python
#크로마 벡터 영구 저장소 지정
persist_directory = '/content/chroma'

#벡터 저장소
vectordb = Chroma.from_documents(
    documents=splits,
    embedding=embedding,
    persist_directory=persist_directory
)

print(vectordb._collection.count())
```

```python
question = "transformer의 주요 작동원리?"

docs = vectordb.similarity_search(question,k=10)  #return 받고자하는 문서의 수를 3개로 지정

# 문서 길이 확인
print(len(docs))

# 첫번째 문서의 콘텐츠 확인
print(docs[0].page_content)

# 영구 저장
vectordb.persist()
```

```python
question = "transformer main keywords five"

docs = vectordb.similarity_search(question,k=5) #return 받고자하는 문서의 수를 5개로 지정

# 처음 2개의 결과 확인
print(docs[0])
print('\n\n', docs[1])
```
